### Seperating X and Y

```python
X = df.drop(columns=['Id','SalePrice'])
Y = df[['SalePrice']]
```

```python
X.head()
```

```python
Y.head()
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
# making pipelines
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)

cat_pipe = make_pipeline(
    SimpleImputer(strategy="constant",fill_value="NotAvailable"),
    OrdinalEncoder()
)
```

```python
# Combining pipelines
# preprocessed data for feature selection

select_pre = ColumnTransformer([
    ("num", num_pipe, con),
    ("cat", cat_pipe, cat)
]).set_output(transform="pandas")
```

```python
x_sel_pre = select_pre.fit_transform(X)
x_sel_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sel = SequentialFeatureSelector(
    LinearRegression(),
    direction="backward",
    n_features_to_select="auto"
)

sel.fit(x_sel_pre,Y)
```

```python
imp_cols = sel.get_feature_names_out()
imp_cols
```

```python
fwdSelCols = [col.split("__")[1] for col in imp_cols]
```

```python
len(fwdSelCols)
```

```python
x_sel = X[fwdSelCols]
```

```python
x_sel.head()
```

```python
cat_sel = list(x_sel.columns[x_sel.dtypes == "object"])
con_sel = list(x_sel.columns[x_sel.dtypes != "object"])
```

```python
cat_sel
```

```python
num_pipe1 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe1 = make_pipeline(
    SimpleImputer(strategy="constant",fill_value="NotAvailable"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False)
)
```

```python
pre2 = ColumnTransformer([
    ("num", num_pipe1, con_sel),
    ("cat", cat_pipe1, cat_sel)
]).set_output(transform="pandas")
```

```python
pre2
```

```python
x_sel_pre1 = pre2.fit_transform(x_sel)
x_sel_pre1
```

```python
pre2
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x_sel_pre1,Y,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain,ytrain)
```

### Linear regression score

```python
model.score(xtest,ytest)
```

### Applying ridge model

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
import numpy as np
```

```python
param = {"alpha": np.arange(0.1, 50, 0.1)}
```

```python
param
```

```python
gscv_ridge = GridSearchCV(Ridge(), param_grid=param, cv=5, scoring="r2")
gscv_ridge.fit(xtrain,ytrain)
```

```python
gscv_ridge.best_score_
```

```python
best_ridge = gscv_ridge.best_estimator_
best_ridge
```

```python
best_ridge.score(xtrain,ytrain)
```

```python
gscv_lasso.best_params_
```

```python
b = best_lasso.score(xtest,ytest)
b
```

# Ridge score better than lasso 

```python
a >= b
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    #predict the model
    ypred = model.predict(x)
    # Calculating the values
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)

    # Printing the values
    print(f"MSE : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 Score : {r2:.4f}")
```

```python
evaluate_model(best_lasso, xtest,ytest)
```

```python
evaluate_model(best_ridge,xtest,ytest)
```
